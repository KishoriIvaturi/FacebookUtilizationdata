### **5.1 Handling of Missing Data**

**Question 4: How many people have zero friends?**

```python
zerofriends = (FBdata['friend_count']== 0).sum()
zerofriends
```

```python
def zerofriends(friend_count):
    if friend_count == 0:
        return 'nill friends'
    else:
        return 'friends'
    
FBdata['zerofriends'] = FBdata['friend_count'].apply(zerofriends)
FBdata['zerofriends'].value_counts()
       
```
